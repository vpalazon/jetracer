```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos
metros_cuadrados = np.array([[10], [60], [70], [90], [110]])
precios = np.array([80, 200, 200, 250, 320])  # En miles de euros

# Crear el modelo y entrenarlo
modelo = LinearRegression()
modelo.fit(metros_cuadrados, precios)

# Predicciones
precios_pred = modelo.predict(metros_cuadrados)

# Coeficientes del modelo
print(f"Pendiente (€/m²): {modelo.coef_[0]:.2f}")
print(f"Intercepción (precio base): {modelo.intercept_:.2f} mil €")

# Predecir el precio para un piso de 80 m²
pred_80m2 = modelo.predict([[80]])
print(f"Precio estimado para 80 m²: {pred_80m2[0]:.2f} mil €")

# Visualización
plt.scatter(metros_cuadrados, precios, color='blue', label='Datos reales')
plt.plot(metros_cuadrados, precios_pred, color='red', label='Línea de regresión')
plt.xlabel('Metros cuadrados')
plt.ylabel('Precio (mil €)')
plt.title('Precio vs Metros Cuadrados')
plt.legend()
plt.grid(True)
plt.show()
```
